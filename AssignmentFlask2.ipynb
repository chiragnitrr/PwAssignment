```python
Q1. Explain GET and POST methods.
Answer : GET and POST are two HTTP methods used to sent the data between a client (such as web browser ) and a server. they define how the
         data is transmitted and purpose of the request. 
        
        GET : The GET method is used to request data from a server . it retrieves information specified in the URLs query parameter.
              Ex : google search is a example of get method.
            
        POST : The POST method is used for sumbit the data to be processed by a server. it sends the data in the body of the HTTP 
               request , rather than appending it to the URL.
```

```python
Q2. Why is request used in Flask?
Answer : In Flask, the 'request' object is used to access incoming request data sent by the client. it allows you to retrive and work with
        data such as form inputs.
    
       some common use cases of request in Flask :
            1. Accessing form data 
            2. Query Parameters
            3. Uploading files
            4. Accessing cookies
            5. Retrieving Headers
```

```python
Q3. Why is redirect() used in Flask?
Answer : In Flask, the 'redirect()' function is used to perform a redirect to a different URL or route. It is handy function that help in
        controlling the flow of the application and redirecting the user to a different page route.
    
    use cases of redirect :
        1.Changing page or Route
        2.Handling Form Submissions
        3.Redirecting After login/logout
        4.Handling Authentication and Authorization
        5.Implementing URL Shortenning
```

```python
Q4. What are templates in Flask? Why is the render_template() function used?
Answer : In Flask, templates are file that contain HTML code mixed with the placeholders for dynamic content. They are used to generate HTML
        pages dynamically based on data and logic provided by Flask application.
    
    uses of 'render_template()' function :
        1.Rendering HTML
        2.Data passing
        3.Template engine integration
```
