```python
#Q2. Write a python program to print Python Exception Hierarchy.

#Answer :
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

# Print the exception hierarchy starting from BaseException
print_exception_hierarchy(BaseException)
```

```python
# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example

# Answer : The ArithmeticError class is the base class for arithmetic related errors. 
#         1.ZeroDivisionError : This error will occur when we divide any number with zero .
#         Example :
            
a = 10
try :
    a/0
    
except ZeroDivisionError as e:
    print(e)
    
         # 2. OverflowError : This error will be occur when result of any arithmetic operation
         #    eceeds it's maximum representable value.
         #    Example :
                
try :
    v = 2**100000

except OverflowError as e:
    print("Error : Numeric overflow")
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
# Answer : The LookupError in the python is the base class for look-up related errors such as
#          KeyError and IndexError .
         
#         KeyError : This error will occur when we will access the element of dictionay with the help of
#                    key which doesn't exist.

my_dictionary = {'a' : 1, 'b' :2, 'c' :3, 'd' :4}
try :
    print(my_dictionary['e'])
          
except KeyError as e:
          print("Key is not present")
        
        # IndexError : This error will occur when we access a sequence (e.g. list, tuple) with the help
        #              index which is out of it's range.
        
my_list = [1,2,3,4,5]
try :
    print(my_list[8])
except IndexError as e:
    print(e)
```
