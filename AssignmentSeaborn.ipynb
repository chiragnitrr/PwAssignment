```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Answer : type of plot use in Seaborn library:
    1. Scatter plot : it used to plot the point data present on different - different axis.
    2. Relation plot : it is type of scatter plot but in arguments we give any relations according to our
                       need for better visualisation.
    3. Catogorical plot : it used to plot the datas of different catogories. in one axis there are catogories
                        and any other parameter on another axis.
    4. Dis plot : it is histogram plot in seaborn.
    5. Joint plot : it is the combination of scatter and dis plot
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.

# Answer :
    
import seaborn as sns
fm = sns.load_dataset('fmri')
```

```python
fm.head()
```

```python
sns.relplot(x = fm.timepoint, y = fm.signal, data = fm, style = 'event', hue = 'region')
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.
# Answer :
    
tn = sns.load_dataset('titanic')
tn.head()
```

```python
sns.boxplot(x = tn.pclass, y = tn.age, data = tn)
```

```python
sns.boxplot(x = tn.pclass, y = tn.fare, data = tn)
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.
# Answer :
    
dm = sns.load_dataset('diamonds')
dm.head()
```

```python
sns.histplot(data = dm, x = 'price', hue = 'cut', kde = True)
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.
# Answer :

ir = sns.load_dataset('iris')
ir.head()
```

```python
sns.pairplot(data = ir, hue = 'species')
```
