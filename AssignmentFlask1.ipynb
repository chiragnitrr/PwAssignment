```python
Q3. What is App routing in Flask? Why do we use app routes?
Answer : in Flask, app routing refers to the process of mapping URL paths to the specific functions or views in
        your application. it allows to define different routes or URLs that user can access in your application
        and associate them with the corresponding python functions that handle the logic  and generate the response.
        
        app routes are used in Flask for the following reasons:
            1. Handling different URL paths
            2. Handling HTTP methods
            3. Creating RESTful APIs
            4. Separating concerns
            5. URL generation
```

```python
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
    
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

Attach the screenshot of the output in Jupyter Notebook.

Answer : 
code :
    
from flask import Flask

myapp = Flask(__name__)

@myapp.route("/")
def hello_world():
    return "<h1>Company Name: ABC Corporation</h1><h1>Location: India</h1><h1>Contact Detail: 999-999-9999</h1>"

@myapp.route("/welcome")
def welcomemsg():
    return "<h1>Welcome to ABC Corporation</h1>"
if __name__ == '__main__':
    myapp.run(host="0.0.0.0")
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Answer : In Flask the 'url_for()' function is used for url building. it is a helper function that generates a
        URL for the specified end point. it takes endpoint name as it first argument and any addition keyword.
  
code : 
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index')) 
        print(url_for('profile', username='John')) 
```
