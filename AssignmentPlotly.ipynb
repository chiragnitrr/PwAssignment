```python
# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
# scatter plot for age and fare columns in the titanic dataset.
```

```python
import plotly.graph_objects as go
import seaborn as sns
```

```python
tn = sns.load_dataset('titanic')
tn.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tn.age, y = tn.fare, mode = 'markers'))
fig.show()
```

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
# Answer :
tp = sns.load_dataset("tips")
tp.head()
```

```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.datasets import load_iris

iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target_names[iris.target]

fig = go.Figure(data=go.Splom(
    dimensions=[dict(label=feat, values=data[feat]) for feat in iris.feature_names],
    marker=dict(color=data['species'],
                colorscale='Viridis',
                showscale=False),))

fig.update_layout(
    title='Scatter Matrix - Iris Dataset',
    height=800,
    width=800,)

fig.show()
```
