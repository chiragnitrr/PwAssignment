Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in 
predicting the quality of wine.

```python
#Answer :
import pandas as pd
df = pd.read_csv("winequalityN.csv")
l = list(df.columns)
print("The key features of the wine quality data set are : ", l)
```

```python
their importance in predicting the quality of red wine:
1. Fixed Acidity: Fixed acidity is the concentration of non-volatile acids in the wine. These acids play a role in determining the 
   overall tartness and acidity level of the wine. They contribute to the wine's structure and balance.
2. Volatile Acidity: Volatile acidity refers to the concentration of volatile acids in the wine. High levels of volatile acidity can
   lead to off-putting aromas and flavors, resembling vinegar. Controlling volatile acidity is crucial for maintaining the wine's 
   pleasantness.
3. Citric Acid: Citric acid is a weak organic acid found in wines. It can enhance the wine's freshness and contribute to its acidity.
   Proper levels of citric acid can add complexity to the wine's flavor profile.
4. Residual Sugar: Residual sugar is the amount of sugar left in the wine after fermentation. It impacts the wine's sweetness and 
   mouthfeel. The balance between residual sugar and acidity is essential for a harmonious taste.
5. Chlorides: Chloride levels affect the wine's saltiness and overall perception. Proper chloride concentration can enhance the wine's
   flavor and mouthfeel, but excessive amounts can lead to undesirable flavors.
6. Free Sulfur Dioxide: Free sulfur dioxide acts as an antioxidant and preservative in wine. It helps prevent spoilage and oxidation,
   which can negatively impact the wine's quality and stability.
7. Total Sulfur Dioxide: Total sulfur dioxide includes both free and bound forms of sulfur dioxide. It's an important measure of the
   wine's ability to resist spoilage and maintain its character over time.
8. Density: Density is a measure of the wine's mass per unit volume. It can provide insights into the wine's alcohol content and 
   overall body. Higher density wines are often more full-bodied.
9. pH: pH measures the acidity or alkalinity of the wine. It influences various chemical reactions that affect taste, color, and 
   stability. Proper pH levels contribute to a balanced and appealing wine.
10. Sulphates: Sulphates are a type of salt that can contribute to the wine's aroma, flavor, and stability. They are often added to
    wines as a preservative.
11. Alcohol: Alcohol content impacts the wine's body, mouthfeel, and overall perception. It can also influence the wine's perceived
    warmth and depth of flavor.
12. Quality (Target Variable): This is the feature you're trying to predict â€“ the quality of the red wine. It's often rated by sensory
    experts based on taste, aroma, balance, and overall enjoyment.
```

Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of 
features. What is the minimum number of principal components required to explain 90% of the variance in 
the data?

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the dataset
# Replace 'wine_data.csv' with the actual path to your dataset
data = pd.read_csv('winequalityN.csv')
data = data.dropna()
# Extract features (X) and labels (y) if applicable
X = data.drop(columns=['quality'])  # Adjust column name based on your dataset
y = data['quality']  # Adjust column name based on your dataset

# Encode categorical variables if present
X = pd.get_dummies(X)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Determine the number of principal components required to explain 90% variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
num_components_90_percent = np.argmax(cumulative_variance_ratio >= 0.9) + 1

print(f"Number of components required to explain 90% variance: {num_components_90_percent}")
```
