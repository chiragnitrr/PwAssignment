```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
Answer : Reality : Null hypothesis is true or Null hypothesis is false.
         Decision : Null hypothesis is true or Null hypothesis is false.
        
        Type 1 error : We reject the Null hypothesis when in reality it is true.
        Example scenario : A medical researcher conducts a clinical trial to test the effectiveness of a new drug for treating
        a cetain disease. The Null hypothesis(H0) is that the new drug has no effect on the disease, while the alternative 
        hypothesis (H1) is that the drug is effective. if the researcher mistakenly conclude that the drug is effective , when 
        it actually has no effect, they commit a type 1 error.
        
        Type 2 error : We retain the Null hypothesis when in reality it is false.
        Example scenario : Continuing with the medical researcher's example, if the new drug actually has a beneficial effect
        on the disease (H1 is true), but the researcher fails to find enough evidence to support this claim and does not reject
        (H0), they commit a Type 2 error.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
Answer : Confience intervals are typically expressed as a range of values along with a confidence level, which is a measure of
         how confident we are that the true population parameter falls within the interval. The most common confidence level is
         95%, but other common choices include 90% and 99%.
        
         To calculate a confidence interval, we need three main components:
         1.Sample statistic
         2.Standard error
         3.Critical value
         
         Confidence Interval = Sample statistic ± Critical Value(CV) * Standard Error(SE)
         
         Example : Suppose we want to estimate the average height of a certain population of adult males. we take a random sample
         of 100 adult males from this population and measure their heights. the sample mean height is 175 cm, and the sample standard
         deviation is 5cm.
         
        Standard error(SE) = (sample standard deviation)/√(sample size) = 5 / √100 = 0.5
        
        For a 95% confidence level,
        Critical value(CV) = 1.96
        Margin of error(ME) = CV * SE = 1.96*0.5 = 0.98
        Confidence Interval = Sample Mean ± Margin of Error
        Confidence Interval = 175 ± 0.98
        The 95% confidence interval for the population mean height is (174.02, 175.98) cm.
```

```python
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and 
    new evidence. Provide a sample problem and solution.
Answer :  Sample Problem : Suppose you have a bag of 100 marbles containing 30 red and 70 blue marbles. you want to know the 
          probability of drawing a red marble from the bag. however after drawing one marble from the bag, you find out that
          it is red. Now, you want to calculate the probability of drawing another red marble from the bag.
          
          A: Drawing a red marble from the bag (prior probability).
          B: Drawing a red marble in the second draw (updated probability with new evidence).
          P(A) = Probability of drawing a red marble in the first draw = 30/100 = 0.3
          P(A|B) = Probabilty of drawing red marble in the second draw, we don't put the first red marble back in the bag.
           
          Bayes' Theorem :
          P(B|A) = (P(A|B)*P(B))/P(A)
          
          P(A) = 0.3
          P(A|B) = 29/98 = 0.29
          P(B) = 30/99 = 0.303
         
          P(B|A) = (0.29*0.303)/0.3 = 0.2997
          so, the probability of drawing a red marble in the second draw, given that the first draw was red, is approximately 
          0.2997 or 29.97%.  
```

```python
Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of 
    a scenario where a larger sample size would result in a smaller margin of error.
Answer : The margin of error (ME) in a confidence interval is a measure of the precision or uncertainty associated with the 
         estimate of a population paramete, such as the population mean or proportion, based on a sample from that popultion. It 
         represents the range around the sample statistic within which we expect the true population parameter to lie with a
         certain level of confidence.
            
         Formula of Margin of error :
        Margin of error(ME) = ((critical value)*(sample standard deviation))/√(sample size)
        We clearly see margin of error is inversely proportion to the sample size if we increase sample size, margin of error decreases
        and we decrese sample size margin of error increases.
        
        Example:
            critical value = 1.96
            sample standard deviation = 5
            
            taking sample size 100
            Margin of error = (1.96 * 5 )/√100 = 0.98
            
            taking sample size 64
            Margin of erro = (1.96 * 5)/√64 = 1.225
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a 
     standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% 
     confidence level using a t-test.
Answer : Null hypothesis : The new weight loss drug is not significantly effective.
         Alternative hypothesis : The new weight loss drug is significantly effective.
    
        t = (x̄ - μ) / (s / √n)
        where :
        x̄ = Sample mean (6 pounds)
        μ = Population mean under the null hypothesis (assuming 0 pounds)
        s = Sample standard deviation (2.5 pounds)
        n = Sample size (50 participants)
        
        t = (6 - 0) / (2.5 / √50)
        t = 6 / (2.5 / 7.07)
        t = 6 / 0.3535
        t = 16.98
        
        degree of freedom (df) = n-1 = 50-1 = 49
        Critical t-value for two-tailed test at a 95% confidence level with 49 df
        critical t-value = ±2.009
        
        using a statical t-table, we find that the p-value is significantly samaller than 0.001 (p<0.001).
        which means we reject the null hypothesis.
```

```python
Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for 
     the true proportion of people who are satisfied with their job.
Answer : Confidence Interval = Sample Proportion ± Margin of Error
         Sample proportion (p̂) = 65% = 0.65
         Sample size (n) = 500
         standard error (SE)  = √((p̂ * (1 - p̂)) / n)
         SE = √((0.65 * (1 - 0.65)) / 500)
         SE ≈ √(0.2275 / 500) ≈ √0.000455 = 0.0213
            
         for a 95% confidence level, two tailed test 
         critical z-value = 1.96
         margin of error (ME) = Critical Z-Value * Standard Error
         ME = 1.96 * 0.0213 = 0.0418
            
        Lower bound = Sample Proportion - ME = 0.65 - 0.0418 = 0.6082
        Upper bound = Sample Proportion + ME = 0.65 + 0.0418 = 0.6918
        
        So, the 95% confidence interval for the true proportion of people who are satisfied with their job is approximately 60.82% 
        to 69.18%.
```

```python
Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score
     of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis 
     test to determine if the two teaching methods have a significant difference in student performance using a t-test with a 
     significance level of 0.01. 
Answer : Null hypothesis : There is no significant difference in student performance between the two teaching methods.
         Alternative hypothesis : There is a significant difference in student performance between the two teaching methods.
        
        The Formula for the indepent samples t-test:
        t = (x̄A - x̄B) / √((sA^2 / nA) + (sB^2 / nB))  
        
        mean score of Sample A (x̄A) = 85
        mean score of Sample B (x̄B) = 82
        standard deviation of Sample A (sA) = 6
        standard deviation of Sample B (sB) = 5
        sample size of Sample (nA) 
        sample size of Sample (nB) 
        taking nA = nB = 30
        
        standard error (SE) = √((sA^2 / nA) + (sB^2 / nB))
        SE = √((6^2 / 30) + (5^2 / 30))
        SE ≈ √(36/30 + 25/30) ≈ √(61/30) = 1.974
        
        t = (85-82)/1.974 = 1.52
        degree of freedom (df) = nA + nB - 2 = 30 + 30 -2 = 58
        For one tailed test at a significance level of 0.01 with 58 df
        critical t-value = 2.658
        
        calculated t-value(1.52) < critical t-value(2.658)
        we fail to reject the null hypothesis.
```

```python
Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 
     90% confidence interval for the true population mean.
Answer : Confidence Interval = Sample Mean ± Margin of Error

         Population mean (μ) = 60
         Population standard deviation (σ) = 8
         Sample size (n) = 50
         Sample mean (x̄) = 65
         Standard Error (SE) = σ/√n = 8/√50 = 1.13137
            
         Critical value for 90% confidence interval, two tail-test
         Critical value (z) = 1.645
            
         Margin of Error (ME) = CV*SE = 1.645*1.13137 = 1.86
        
         Lower bound = Sample Mean (x̄) - ME = 65 - 1.86 = 64.14
         Upper bound = Sample Mean (x̄) + ME = 65 + 1.86 = 66.86
        
         90% confidence interval for the true population mean is approximately 63.14 to 66.86.
```
