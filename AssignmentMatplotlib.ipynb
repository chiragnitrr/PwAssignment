```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("this is title")
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#Answer : The subplot fuction is used to plot multiple graphs, scatter, bar or histogram in a single figure.

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(3,2,1)
ax1.plot([0, 1, 2, 3, 4, 5], [0, 100, 200, 300, 400, 500])

fig = plt.figure()
ax1 = fig.add_subplot(3,2,1)
ax1.plot([0, 1, 2, 3, 4, 5], [50, 20, 40, 20, 60, 70])

fig = plt.figure()
ax1 = fig.add_subplot(3,2,1)
ax1.plot([0, 1, 2, 3, 4, 5], [10, 20, 30, 40, 50, 60])

fig = plt.figure()
ax1 = fig.add_subplot(3,2,1)
ax1.plot([0, 1, 2, 3, 4, 5], [200, 350, 250, 550, 450, 150])

plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

#Answer : Bar plot is type of visual representation of data, it used to compair between different - different datas of
#          any product, company or anything.
import matplotlib.pyplot as plt
plt.barh(company,profit)
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
                        
# Answer : A box plot also known as box-and-whisker plot, is graphical representation of dataset that provide the
#          summery of its distribution. it displays the five-number summary of a data set which is minimum, first
#          quartile(25 percentile), median (50 percentile), third quartile (75 percentile) and maximum. it used for
#         summarizing the data and compairing the distribution.
                        
data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot')
plt.show()
```
