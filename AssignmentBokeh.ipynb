```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
# Answer : In Bokeh, glyphs are visual makers used to represent data points in a plot. They can be
#          various shapes, such as circle, squares, triangles, or lines, and they can be customized
#          with different attributes like size, color, or transparency. Glyphs are added to a bokeh
#          plot using glyph methods provided by the 'figure' object.
#          Example:
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
x = [1,2,3,4,5]
y = [6,7,2,4,5]
p = figure()
p.circle(x,y, size=10, color = 'blue',alpha = 0.6)
p.line(x,y, line_width = 2, color = 'red')
show(p)
```
